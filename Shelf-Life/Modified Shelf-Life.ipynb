```python
import pandas as pd
import numpy as np
import time
import random
```

```python
#Csv file to get the expiry daye of each food-items
shelf = pd.read_csv("shelf_best.csv")
```

```python
shelf.head(20)
```

```python
#it will store all the food items of all customers with a counter variable
"""
I have used python for this task.
You can re-implement it in javascript along with making database to store
the data in customer table and the shelf-life table
Here I have used dictionary.
"""
customer=dict()

#Creating random Dummy dataset
for i in range(100):
    customer[i]={}
    for item in random.choices(shelf.name.to_list(),k=10):

        exp_days=shelf[shelf.name==item]["min_day"].values[0]

        customer[i][item]=random.randint(int(0.4*exp_days),int(1.2*exp_days))
```

```python
#expiray time of all items
"""
shelf-life to store the expiry date of each item. 
"""
shelf_life={}
for index, row in shelf.iterrows():
    if type(row['min_day'])==int:
        shelf_life[row['name']] = row['min_day']
```

```python
#this function will be called whenever  a customer buy something
#It will add items if not in the table else reset the counte value
#it will add a new customer in the dict
"""
for roduction purpose you have to store the data into a database using NodeJS
and then re-implementing the foloowing function.
"""
def update(customer_idx,items):
    if not customer_idx in customer.keys():
        customer[customer_idx]={}
        
    for item in items:
            customer[customer_idx][item]=0
    print(f'{", ".join(items)} has been added in the database')
    print("------------------")
    print("The current items list:")
    print(customer[customer_idx])
```

```python
print("Customer-1 shelf list, Before updating:")
print(customer[1])
print()
print("After updating")
update(1,["peanut","mango"])
```

```python
#this function will be called everyday to increase the counter
#if counter value more than expiray date then it will delete the items in the table 
def counter():
    c=0
    for idx in list(customer.keys()):
        for item in list(customer[idx].keys()):
            customer[idx][item]+=1
            if shelf_life.get(item,9999)<customer[idx][item]:
                c+=1
                print(f"id-{idx} {item} has been expired thereby removing it from the shelf")
                del customer[idx][item]
    print()
    print(f"Total: {c} items has been removed from the shelf")
#     print(len(customer))
```

```python
print("Customer-1 shelf list, Before counter:")
print(customer[1])
print()
print("After applying counter")
counter()
```

```python
#this function will be called by api everyday and when the custmer buys something
def main(isupdate=False,iscounter=False,idx=None,items=None):
    if isupdate:
        update(idx,items)
        
    elif iscounter:
        counter()
```
